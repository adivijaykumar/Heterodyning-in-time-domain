```python
def SNR_using_bilby(fiducial_parameters,f_min = 10):
    duration = 256.0
    sampling_frequency = 4096.0
    minimum_frequency = f_min

    injection_parameters2 = fiducial_parameters.copy()
    injection_parameters2["mass_1"], injection_parameters2["mass_2"] = (
        chirp_mass_and_mass_ratio_to_component_masses(
            injection_parameters2["chirp_mass"], injection_parameters2["mass_ratio"]
        )
    )
    injection_parameters2.pop("chirp_mass")
    injection_parameters2.pop("mass_ratio")
    print(injection_parameters2["mass_1"]+injection_parameters2["mass_2"])

    # Fixed arguments passed into the source model
    waveform_arguments = dict(
        waveform_approximant="IMRPhenomT",
        reference_frequency=minimum_frequency,
        minimum_frequency=minimum_frequency,
    )

    # Create the waveform_generator using a LAL BinaryBlackHole source function
    waveform_generator = bilby.gw.WaveformGenerator(
        duration=duration,
        sampling_frequency=sampling_frequency,
        frequency_domain_source_model=bilby.gw.source.lal_binary_black_hole,
        parameter_conversion=bilby.gw.conversion.convert_to_lal_binary_black_hole_parameters,
        waveform_arguments=waveform_arguments,
        parameters=injection_parameters2,
    )

    # Set up interferometers.  In this case we'll use two interferometers
    # (LIGO-Hanford (H1), LIGO-Livingston (L1). These default to their design
    # sensitivity
    ifos = bilby.gw.detector.InterferometerList(["ET","CE"])
    ifos[0].minimum_frequency = f_min
    ifos[1].minimum_frequency = f_min
    ifos[2].minimum_frequency = f_min
    ifos[3].minimum_frequency = f_min
    ifos.set_strain_data_from_zero_noise(
        sampling_frequency=sampling_frequency,
        duration=duration,
        start_time=fiducial_parameters["geocent_time"],
    )
    ifos.inject_signal(
        waveform_generator=waveform_generator, parameters=injection_parameters2
    )
    likelihood = bilby.gw.likelihood.GravitationalWaveTransient(
        interferometers=ifos,
        waveform_generator=waveform_generator
    )
    return {
        "ET1": ifos.meta_data["ET1"]["matched_filter_SNR"].real,
        "ET2": ifos.meta_data["ET2"]["matched_filter_SNR"].real,
        "ET3": ifos.meta_data["ET3"]["matched_filter_SNR"].real,
        "CE": ifos.meta_data["CE"]["matched_filter_SNR"].real
    }, np.sqrt(
        (ifos.meta_data["ET1"]["matched_filter_SNR"].real) ** 2
        + (ifos.meta_data["ET2"]["matched_filter_SNR"].real) ** 2
        + (ifos.meta_data["ET3"]["matched_filter_SNR"].real) ** 2
    + (ifos.meta_data["CE"]["matched_filter_SNR"].real) ** 2),likelihood
```
