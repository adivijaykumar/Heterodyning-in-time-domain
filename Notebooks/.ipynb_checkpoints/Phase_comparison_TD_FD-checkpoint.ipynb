```python
import numpy as np
import matplotlib.pyplot as plt
from pycbc.waveform import get_td_waveform, get_fd_waveform, utils
```

```python
for apx in ['IMRPhenomD']:
    hp_td, hc_td = get_td_waveform(approximant=apx,
                                 mass1=10,
                                 mass2=10,
                                 delta_t=1.0/4096,
                                 f_lower=40)
h_td = hp_td-1j*hc_td
```

```python
hp_td.sample_times.data
```

```python
for apx in ['IMRPhenomD']:
    hp_fd, hc_fd = get_fd_waveform(approximant=apx,
                                 mass1=10,
                                 mass2=10,
                                 delta_t=1.0/4096,
                                 delta_f = 1/32,
                                 f_lower=40)
```

```python
plt.plot(hp_fd)
plt.xlim(0,5000)
# plt.xscale("log")
# plt.yscale('log')
```

```python
plt.plot(hp_td)
plt.xlim(44500,45500)
```

```python
plt.plot(utils.phase_from_polarizations(hp_td, hc_td))
plt.plot(np.unwrap(np.angle(hp_fd)))
```
