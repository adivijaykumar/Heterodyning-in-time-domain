```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Time_array = np.arange(-2.0+0.5,0.5,1/4096)
```

```python
def BinsUsingTimeGrid(time_grid):
    return [time_grid[0],time_grid[-1]]
```

```python
def BisectBins(tolperbin,time_grid):
    if LikeError(time_grid) < tolperbin or len(time_grid) <= 2:
        return BinsUsingTimeGrid(time_grid)
    else:
        mid = len(time_grid) // 2
        left_time_grid = time_grid[:mid+1]
        right_time_grid = time_grid[mid:]
        left_bins = BisectBins(tolperbin,left_time_grid)
        right_bins = BisectBins(tolperbin,right_time_grid)
        return left_bins[:-1] + right_bins
```

```python
def LikeError(time_bin):
    return np.abs(time_bin[-1]-time_bin[0])
```

```python
def SetBins(tolperbin,Time_array):
    time_bins = np.unique(np.array(BisectBins(tolperbin,list(Time_array))))
    Bin_edge_index = np.searchsorted(Time_array, time_bins)
    number_of_bins = len(time_bins) - 1
    return time_bins, Bin_edge_index, number_of_bins
```

```python
SetBins(0.1,Time_array)
```
